## TOWER OF HANOI

#### The Tower of Hanoi is a classic mathematical puzzle that involves moving a stack of disks from one peg to another peg with the help of a third peg. The objective of the puzzle is to move the entire stack of disks from the source peg to the destination peg, without placing a larger disk on top of a smaller disk.

#### Here is a step-by-step explanation of the algorithm:

1. Begin by labeling the pegs as A, B, and C.

2. Number the disks from 1 to n, with the smallest disk being 1 and the largest disk being n.

3. Initially, all the disks are stacked on peg A in order of size, with the smallest disk on top and the largest disk on the bottom.

4. To solve the puzzle, we recursively move the top n-1 disks from peg A to peg B using peg C as an intermediate peg, and then move the remaining largest disk from peg A to peg C.

5. Finally, we recursively move the n-1 disks from peg B to peg C using peg A as an intermediate peg.

Use case:

#### The Tower of Hanoi algorithm is often used as an exercise in computer science and programming courses to teach recursion and problem-solving.
#### It is also used in real-world applications, such as in the optimization of robotic arm movements and in the design of communication networks.

```python
# This function implements the recursive solution to the Tower of Hanoi problem
# The Tower of Hanoi is a classic algorithmic problem that involves moving a stack
# of disks from the first tower to the third tower while obeying certain rules.

def tower_of_hanoi(num_disks, start_tower, mid_tower, end_tower):
  
  # Base case: if there's only one disk, move it from start_tower to end_tower
  if(num_disks == 1):
    print("Move Disk 1 from", start_tower, "to", end_tower)
    return
  
  # Move the top n-1 disks from the start_tower to the mid_tower using the end_tower as the intermediate
  tower_of_hanoi(num_disks - 1, start_tower, end_tower, mid_tower)
  
  # Move the nth disk from start_tower to end_tower
  print("Move disk", num_disks, "from ", start_tower, "to ", end_tower)
  
  # Move the n-1 disks from the mid_tower to the end_tower using the start_tower as the intermediate
  tower_of_hanoi(num_disks - 1, mid_tower, start_tower, end_tower)


# Program start
print("Starting Tower is A \nFinal Tower is C\n")

# Get input for number of disks
num_disks = int(input("Enter no. of Disks : "))

# Call the function to solve the Tower of Hanoi problem
tower_of_hanoi(num_disks, 'A', 'B', 'C')
```
